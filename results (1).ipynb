```python
import matplotlib.pyplot as plt
import numpy as np

#  
data = {
    'ANN': {'time': 91.74, 'mae': 0.5360, 'mse': 0.6868},
    'CNN': {'time': 145.39, 'mae': 0.6892, 'mse': 0.9561},
    'RNN': {'time': 896.55, 'mae': 0.8569, 'mse': 1.3992}
}

#  

colors = ['#2ecc71', '#3498db', '#e74c3c']
models = list(data.keys())

#   
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 15), dpi=100)

# 1.   
times = [data[model]['time'] for model in models]
bars = ax1.bar(models, times, color=colors, alpha=0.8)
ax1.set_title('   ', fontsize=14, pad=20)
ax1.set_ylabel(' ()', fontsize=12)
ax1.tick_params(axis='both', which='major', labelsize=10)
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.1f}',
            ha='center', va='bottom')

# 2.   MAE
mae_values = [data[model]['mae'] for model in models]
width = 0.4
x = np.arange(len(models))
bars1 = ax2.bar(x - width/2, mae_values, width, label='MAE', color=colors, alpha=0.8)

# 3.   MSE
mse_values = [data[model]['mse'] for model in models]
bars2 = ax2.bar(x + width/2, mse_values, width, label='MSE', color=colors, alpha=0.6)

#   
ax2.set_title('   ', fontsize=14, pad=20)
ax2.set_ylabel(' ', fontsize=12)
ax2.set_xticks(x)
ax2.set_xticklabels(models)
ax2.legend(fontsize=10)
ax2.tick_params(axis='both', which='major', labelsize=10)

#    
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.3f}',
                ha='center', va='bottom', fontsize=9)

# 4.   ()
metrics = ['mae', 'mse']
x = np.arange(len(metrics))  #  
width = 0.25  #  

for i, model in enumerate(models):
    values = [data[model]['mae'], data[model]['mse']]
    ax3.bar(x + (i*width), values, width, label=model, color=colors[i])

#   
ax3.set_title('    ', fontsize=14, pad=20)
ax3.set_ylabel('', fontsize=12)
ax3.set_xticks(x + width)
ax3.set_xticklabels(['MAE', 'MSE'])
ax3.legend(fontsize=10)
ax3.tick_params(axis='both', which='major', labelsize=10)

#  
plt.tight_layout(pad=3.0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#  
data = {
    'ANN': {'time': 21.24, 'mae': 565.57, 'mse': 618755.38},
    'CNN': {'time': 25.76, 'mae': 881.07, 'mse': 1257046.38},
    'RNN': {'time': 165.99, 'mae': 2641.70, 'mse': 8534947.00}
}

#  

plt.rcParams['font.family'] = 'DejaVu Sans'
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
models = list(data.keys())

#   
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 16), dpi=100)

# 1.    ( )
times = [data[model]['time'] for model in models]
bars = ax1.bar(models, times, color=colors, alpha=0.8, edgecolor='black')
ax1.set_title('  ', fontsize=14, pad=15, fontweight='bold')
ax1.set_ylabel(' ()', fontsize=12)
ax1.set_yscale('log')  #    
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height*1.05,
            f'{height:.2f}',
            ha='center', va='bottom',
            fontsize=10,
            bbox=dict(facecolor='white', alpha=0.8, edgecolor='none'))

# 2.   (MAE  MSE)
metrics = ['mae', 'mse']
metric_labels = ['MAE', 'MSE']
x = np.arange(len(models))
width = 0.35

for i, metric in enumerate(metrics):
    offset = width * i
    values = [data[model][metric] for model in models]
    rects = ax2.bar(x + offset, values, width, label=metric_labels[i], 
                   color=colors[i], alpha=0.8, edgecolor='black')
    for rect in rects:
        height = rect.get_height()
        ax2.text(rect.get_x() + rect.get_width()/2., height*1.02,
                f'{height:,.0f}' if metric == 'mse' else f'{height:.1f}',
                ha='center', va='bottom',
                fontsize=9,
                rotation=45 if metric == 'mse' else 0)

ax2.set_title('  ', fontsize=14, pad=15, fontweight='bold')
ax2.set_xticks(x + width/2)
ax2.set_xticklabels(models)
ax2.legend(fontsize=12)
ax2.grid(True, which='both', linestyle='--', linewidth=0.5)
ax2.set_yscale('log')  




#  
plt.tight_layout(pad=4.0)
plt.show()

```
